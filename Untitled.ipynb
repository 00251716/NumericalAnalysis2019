```python
import numpy as np
import matplotlib.pyplot as plt

g = 32.17
k = 0.1
m = 0.25
s0 = 300

def s(t):
    return s0 - (m*g*t)/k + (((m**2)*g)/k**2)*(1 - np.exp( (-k*t)/m ))
```

```python
t = np.arange(5, 7, 0.1)
plt.grid(True)
plt.plot(t, s(t))
```

```python
print(s(0))
print(s(10))
```

```python
def f(x):
    return np.exp(-0.2*x) - np.exp(-0.75*x) - 1/4

x = np.arange(0.0, 2.0, 0.1)
plt.grid(True)
plt.plot(x, f(x))
```

```python
def g1(x):
    return (-1/0.2)*np.log(1/4 + np.exp(-0.75*x))
```

```python
plt.plot(x, g1(x), 'bo', x, x, 'k')
```

```python
def g2(x):
    return (-1/0.75)*(np.log(np.exp(-0.2*x) - 1/4))
```

```python
plt.plot(x, g2(x), 'bo', x, x, 'k')
```

```python
print(g1(0))
print(g1(2))
print(g2(0))
print(g2(2))


g2(0.5)
```
